```python
xmin = 115
xmax = 350
ymin = 210
ymax = 445
```

```python
class PrepareFrames:
    '''Класс осуществляет извлечение и обработку каждого 15 кадра из видео'''


    def get_frames(self, initial_index, final_index):
        '''Принимает на вход номера начального и коненчого кадров
        и возвращает каждый 15 кадр в этом интервале'''
        
        frames = []
        frame_count = initial_index
        while True:
            ret, frame = cap.read()
            if ret is False:
                break
            if frame_count % 15 == 0:
                frames.append(frame)
            frame_count += 1
            if frame_count % 10000 == 0:
                print(frame_count)
            if frame_count > final_index:
                initial_index = frame_count
                final_index = final_index + 40000
                break
          
        return frames, initial_index, final_index

    def crop_image(self, serie, frames):
        '''Принимает на вход номер серии поступивших кадров, 
        обрезает кадры по координатам, 
        изменяет их размер,
        записывает обрезанные кадры на диск'''
        
        for i, image in enumerate(frames):
            crop_image = image[ymin:ymax, xmin:xmax]
            resized_image = cv2.resize(crop_image, (119, 119))
            cv2.imwrite(f'{serie}_{i}.jpg', resized_image)

    def prepare_frames(self, initial_index, final_index, serie):
        '''Реализует цикл обработки каждого 15го кадра в интервале 40000 кадров'''

        frames, initial_index, final_index = self.get_frames(initial_index, final_index)
        self.crop_image(serie, frames)
        serie += 1
        return serie, initial_index, final_index
```
